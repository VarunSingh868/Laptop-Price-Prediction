```python
from bs4 import BeautifulSoup
import requests
```

```python
file = open('laptoplinks2.txt')
allLinks = file.readlines()
print(allLinks)
print(len(allLinks))
```

```python
domainURL = 'https://www.flipkart.com'
pageURL = domainURL + allLinks[1]
print(pageURL)
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}
response = requests.get(pageURL,headers=headers)
print(response)
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(response.text)
```

```python
def extract_details(response):    
    soup = BeautifulSoup(response.text, 'html.parser')
    details = {}
    
    # Extracting basic details
    details['brand'] = soup.find('span', class_='B_NuCI').get_text(strip=True).split()[0]
    details['price'] = soup.find('div', class_='_30jeq3 _16Jk6d').get_text(strip=True)
    details['type'] = soup.find('td', string='Type').find_next_sibling('td').text.strip()
    details['colour'] = soup.find('td', string='Color').find_next_sibling('td').text.strip()
    
    # Set default value 'No' for 'MS Office Provided'
    details['msoffice'] = 'No'
    msoffice_tag = soup.find('td', string='MS Office Provided')
    if msoffice_tag is not None:
        details['msoffice'] = msoffice_tag.find_next_sibling('td').text.strip()

    # Extracting Memory details
    details['RAM'] = soup.find('td', string='RAM').find_next_sibling('td').text.strip()
    details['RAMtype'] = soup.find('td', string='RAM Type').find_next_sibling('td').text.strip()
    details['SSD'] = soup.find('td', string='SSD').find_next_sibling('td').text.strip()
    if details['SSD'] == 'Yes':
        details['SSDCap'] = soup.find('td', string='SSD Capacity').find_next_sibling('td').text.strip()
    else:
        details['SSDCap'] = 'No'

    HDDCap_tag = soup.find('td', string='HDD Capacity')
    if HDDCap_tag is not None:
        details['HDDCap'] = HDDCap_tag.find_next_sibling('td').text.strip()
    else:
        details['HDDCap'] = 'No'

    # Extracting GPU and CPU
    details['ProcessorBrand'] = soup.find('td', string='Processor Brand').find_next_sibling('td').text.strip()
    details['ProcessorName'] = soup.find('td', string='Processor Name').find_next_sibling('td').text.strip()
    details['ProcessorVariant'] = None
    processor_variant_tag = soup.find('td', string='Processor Variant')
    if processor_variant_tag:
        details['ProcessorVariant'] = processor_variant_tag.find_next_sibling('td').text.strip()
    
    details['ClockSpeed'] = None
    clock_speed_tag = soup.find('td', string='Clock Speed')
    if clock_speed_tag:
        details['ClockSpeed'] = clock_speed_tag.find_next_sibling('td').text.strip()

    details['GPU'] = soup.find('td', string='Graphic Processor').find_next_sibling('td').text.strip()

    GPUramtypetag = soup.find('td', string='Dedicated Graphic Memory Type')
    if GPUramtypetag is not None:
        details['GPUramtype'] = GPUramtypetag.find_next_sibling('td').text.strip()
    else:
        details['GPUramtype'] = 'GDDR4'

    GPUramcaptag = soup.find('td', string='Dedicated Graphic Memory Capacity')
    if GPUramcaptag is not None:
        details['GPUramcap'] = GPUramcaptag.find_next_sibling('td').text.strip()
    else:
        details['GPUramcap'] = '2 GB'

    # Extracting Display and OS
    details['OS'] = soup.find('td', string='Operating System').find_next_sibling('td').text.strip()

    Touchscreentag = soup.find('td', string='Touchscreen')
    if Touchscreentag is not None:
        details['Touchscreen'] = Touchscreentag.find_next_sibling('td').text.strip()
    else:
        details['Touchscreen'] = 'No'

    details['Screensize'] = soup.find('td', string='Screen Size').find_next_sibling('td').text.strip()
    details['Screenreso'] = soup.find('td', string='Screen Resolution').find_next_sibling('td').text.strip()

    refreshratetag = soup.find('td', string='Refresh Rate')
    if refreshratetag is not None:
        details['refreshrate'] = refreshratetag.find_next_sibling('td').text.strip()
    else:
        details['refreshrate'] = '60'

    # Extracting additional details
    details['Bluetooth'] = soup.find('td', string='Bluetooth').find_next_sibling('td').text.strip()
    
    # Set default value None for 'Weight'
    details['Weight'] = None
    weight_tag = soup.find('td', string='Weight')
    if weight_tag:
        details['Weight'] = weight_tag.find_next_sibling('td').text.strip()

    # Set default value None for 'BacklitKB'
    details['BacklitKB'] = None
    BacklitKBtag = soup.find('td', string='Backlit Keyboard')
    if BacklitKBtag is not None:
        sibling = BacklitKBtag.find_next_sibling('td')
        if sibling:
            details['BacklitKB'] = sibling.text.strip()
    else:
        details['BacklitKB'] = 'No'

    details['warranty'] = soup.find('td', string='Warranty Summary').find_next_sibling('td').text.strip()

    return details
```
