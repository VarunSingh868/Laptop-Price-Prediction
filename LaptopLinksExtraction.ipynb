## Extracting Page Links
#### Importing required Libraries
***

```python
from bs4 import BeautifulSoup
import requests
```

#### Declaring URL and headers and trying for a single page
***

```python
url = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US, en;q=0.5'
}

response = requests.get(url, headers=headers)
```

#### Checking response Successful or not:
- if response = 200: successful
- else: fail
***

```python
print(response)
```

- Response successful checking the data extracted in HTML format
***

```python
print(response.text)
```

#### Extracting All callable pages links and storing them in a file
***

```python
file = open('laptoplinks2.txt','w')
for i in range(1,42):
    URL = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=' + str(i)
    r = requests.get(URL,headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    links = soup.find_all('a', class_='_1fQZEK')  
    for link in links:
        href = link.get('href')  
        file.write(href + '\n')
        print(href)
```
